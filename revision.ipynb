```python
data = pd.read_csv('owid-covid-data.csv')


data = data[['iso_code','continent','new_deaths','total_deaths','new_cases','location','total_cases','date','total_vaccinations']]
data = pd.DataFrame(data)
```

```python
print(data['continent'].value_counts())
print(data['continent'].unique())
```

```python
# now it is time to filtering
con = ['UZB']
filt = data['iso_code'].isin(con)
filt
```

```python
a = data[filt]
```

```python
plt.figure(figsize=(14,8))
sns.set()
plt.plot(a['total_cases'])
plt.xticks(fontweight='bold')
plt.yticks(fontweight='bold')
plt.show()
```

```python
# now we will display pie 
val = np.sum(a['new_cases']),np.sum(a["new_deaths"])
lab = ("total cases",'total deaths')
```

```python
plt.figure(figsize=(19,10))
plt.pie(val,labels=lab,autopct='%.2f%%',shadow=True,explode=[0,0.4],textprops={
    'fontweight':"bold",
    'size':'13'
})

plt.title('Uzbekistan covid info',fontweight='bold',fontsize=17)
```

```python
africa = ['Africa']
asia = ['Asia']
europe = ['Europe']
n_amerika = ['North America']
s_amerika = ['South America']
oceania = ['Oceania']
```

```python
afric = data['continent'].isin(africa)
asi = data['continent'].isin(asia)
euro = data['continent'].isin(europe)
nameric = data['continent'].isin(n_amerika)
sameric = data['continent'].isin(s_amerika)
oceani = data['continent'].isin(oceania)
```

```python
afric = data[afric]
asi = data[asi]
euro = data[euro]
nameric = data[nameric]
sameric = data[sameric]
oceani = data[oceani]
vlues = [np.sum(afric['new_cases']),np.sum(asi['new_cases']),np.sum(euro['new_cases']),np.sum(nameric['new_cases']),
        np.sum(sameric['new_cases']),np.sum(oceani['new_cases'])]
labs = ('Afrika','Asia','Europe','North America','South America',"Oceania")
```

```python
plt.figure(figsize=(19,10))
plt.pie(vlues,labels=labs,autopct="%.2f%%",shadow=True,textprops={
    'rotation':4,
    'size':12,
    'fontweight':'bold',
    'color':'black'
},
        explode=[0,0.08,0,0,0,0]
       )
plt.legend(fontsize=13)
plt.title('Covid info: according to continents. Total number of covid cases in persentage.',fontsize=15,fontweight='bold')
plt.show()
```
